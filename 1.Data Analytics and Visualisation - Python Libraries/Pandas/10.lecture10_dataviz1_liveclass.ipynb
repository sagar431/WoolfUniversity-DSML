# How can we find the top-5 genres?

```python
cat_counts = data['Genre'].value_counts()
cat_counts
```

```python
type(cat_counts)
```

```python
x_bar = cat_counts.index
y_bar = cat_counts
```

```python
x_bar
```

```python
plt.bar(x_bar, y_bar)
```

```python
# How can we handle overlapping labels?

# Decrease the font size (not preferred)
# Increase the figure size
# Rotate the labels
```

```python
plt.figure(figsize = (12,8))
plt.bar(x_bar, y_bar)
```

```python
plt.figure(figsize = (12,8))
plt.bar(x_bar, y_bar)
plt.xticks(rotation=90, fontsize=12)
```

```python
plt.figure(figsize = (12,8))
plt.bar(x_bar, y_bar, width = 0.4)
plt.yticks(np.arange(100, 4000))
plt.xticks(rotation=90, fontsize=12)
```

```python
plt.figure(figsize=(10,6))
plt.bar(x_bar,y_bar,width=0.2,color='orange')
plt.title('Games per Genre',fontsize=15)
plt.xlabel('Genre',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation = 90, fontsize=12)
plt.yticks(fontsize=12)
```

```python
plt.figure(figsize=(10,6))
plt.bar(x_bar,y_bar,width=0.2,color='orange')
plt.title('Games per Genre',fontsize=15)
plt.xlabel('Genre',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation = 90, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize = (12,6))
sns.countplot(x="Genre", data = data,order= data['Genre'].value_counts().index, color = 'cornflowerblue')
plt.xticks(rotation = 90)
plt.show()
```

## Pie Chart

```python
sales_data = data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]

region_sales = sales_data.T.sum(axis='columns')
```

```python
region_sales
```

```python
plt.pie(region_sales,
       labels=region_sales.index,
       startangle=90,
      explode=(0.2,0,0,0),
       autopct = '%.3f%%') # label the wedges with their numeric value
plt.show()
```

## Box plot

```python
plt.figure(figsize=(10,10))
sns.boxplot(y = data["Global_Sales"], width=0.2)
plt.yticks(fontsize=20)
plt.ylabel('Global Sales (in million dollars)', fontsize=20)
plt.title('Global Sales of video games', fontsize=20)
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x = data["Genre"], y = data["Global_Sales"], width=0.5)
plt.yticks(fontsize=20)
plt.ylabel('Global Sales (in million dollars)', fontsize=20)
plt.title('Global Sales of video games', fontsize=20)
```

```python
data[data['Genre'] == 'Adventure']['Platform']
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x = data[data['Genre'] == 'Adventure']['Platform'],
            y = data[data['Genre'] == 'Adventure']["Global_Sales"],
            width=0.5)

plt.yticks(fontsize=20)
plt.ylabel('Global Sales (in million dollars)', fontsize=20)
plt.title('Global Sales of video games', fontsize=20)
```
